```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
from itertools import product, combinations
c1 = '#1f77b4'
c1b = 'blue'
c2 = '#ff7f0e'
c2b = 'red'
gr = '#7f7f7f'
```

```python
#Appartenenza a Numeri molto più grandi di 1
%matplotlib inline
x=[-50,100]
y=[-0.006,-0.006]
x1=[102, 200]
y1=[0.994,0.994]
plt.step(x, y, color=c1, linewidth=2.5)
line_cyan, = plt.step(x1, y1, color=c1, label="Crisp", linewidth=2.5)
x1=[-50,5,100,200]
y1=[0.006,0.006,1.006,1.006]
line_orange, = plt.plot(x1,y1, color=c2, label="Fuzzy",linewidth=2.5)
plt.legend(handles=[line_cyan, line_orange])
plt.xlabel("Numeri Reali")
plt.ylabel("Grado di appartenenza")
plt.savefig("fuzzycrisp.pdf", bbox_inches = 'tight', pad_inches = 0)
```

```python
#Appartenenza all'insieme Persone Giovani
%matplotlib inline
def young(x):
    r = 2*math.exp((math.log(0.5)/50)*x) -1
    if r>0:
        return r
    else:
        return 0

cx=np.linspace(0,54,12)
cx=cx[:-1]
for x in cx:
    plt.plot([x,x+3],[young(x),young(x)], color=c1, linewidth=3)
plt.xlabel("Eta\' in anni")
plt.ylabel("Grado di appartenenza")
plt.savefig("giovani.pdf", bbox_inches = 'tight', pad_inches = 0)
```

```python
#Bevande tiepide
%matplotlib inline
mu = 40
variance = 50
sigma = math.sqrt(variance)
x = np.linspace(0, 80, 500)
plt.plot(x,mlab.normpdf(x, mu, sigma)*15, color=c1, linewidth=3)
plt.xlabel("Gradi centigradi")
plt.ylabel("Grado di appartenenza")
plt.savefig("tiepide.pdf", bbox_inches = 'tight', pad_inches = 0)
```

```python
#Margine svm
def produce1(n):
    x=[]
    y=[]
    i=0
    while(i<n):
        cx=10*np.random.random()
        cy=10*np.random.random()
        if(cy>cx+3.5):
            x.append(cx)
            y.append(cy)
            i=i+1
    return x, y
            
def produce2(n):
    x=[]
    y=[]
    i=0
    while(i<n):
        cx=10*np.random.random()
        cy=10*np.random.random()
        if(cy<cx-3.5):
            x.append(cx)
            y.append(cy)
            i=i+1
    return x, y
            
fig, ax = plt.subplots()  
s=60
plt.plot([0,10],[0,10], linewidth=3, color=gr)
plt.plot([0,10],[2,12], linewidth=1.5, color=gr)
plt.plot([0,10],[-2,8], linewidth=1.5, color=gr)
x1=[4.5]
y1=[6.5]
x2=[3.2]
y2=[1.2]
plt.scatter(x1,y1,color=c1, s=s)
plt.scatter(x2,y2,color=c2, s=s, marker="s")
x, y = produce1(15)
plt.scatter(x,y,color=c1, s=s)
x, y = produce2(15)
plt.scatter(x,y,color=c2, s=s, marker="s")
plt.xlim(-1,15)
#plt.arrow(4,6,-0.5,-2.5)
#plt.arrow(3.2,1.2,-1.2,+0.8)
#plt.text(2,-1.5,"y=x-1", fontsize=12)
#plt.text(7,10.5,"y=x+1", fontsize=12)
plt.text(10.5,9.5,"$wx + b = 0$", fontsize=12)
ax.axis('off')
plt.savefig("margine.pdf", bbox_inches = 'tight', pad_inches = 0)
```

```python
#fuzzification
#Margine svm
def produce5(n):
    x=[]
    y=[]
    i=0
    while(i<n):
        cx=4*np.random.random()
        cy=4*np.random.random()
        if np.random.random() > 0.5:
            cx=-cx
        if np.random.random() > 0.5:
            cy=-cy
        if(cx**2 + cy**2 < 2):
            x.append(cx)
            y.append(cy)
            i=i+1
    return x, y
            
def produce6(n):
    x=[]
    y=[]
    i=0
    while(i<n):
        cx=4*np.random.random()
        cy=4*np.random.random()
        if np.random.random() > 0.5:
            cx=-cx
        if np.random.random() > 0.5:
            cy=-cy
        if((cy>=-cx+0.8) or (cy<cx+0.8)):
            x.append(cx)
            y.append(cy)
            i=i+1
    return x, y
fig = plt.figure(figsize=(4,4))
circle1=plt.Circle((0,0),1.5,color=c1,alpha=0.4,linewidth=3)
plt.gcf().gca().add_artist(circle1)
plt.xlim(-4,4)
plt.ylim(-4,4)
x, y = produce5(40)
plt.scatter(x,y,color=c2)
x, y = produce6(15)
plt.scatter(x,y,color=c2)
plt.xlabel("$\phi(x)^{(1)}$", fontsize=12)
plt.ylabel("$\phi(x)^{(2)}$", fontsize=12)
plt.savefig("fuzzification11.pdf", bbox_inches = 'tight', pad_inches = 0)
```
