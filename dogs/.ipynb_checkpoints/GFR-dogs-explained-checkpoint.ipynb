Carico i dati

```python
import numpy as np
import pandas as pd

from possibilearn import *
from possibilearn.kernel import GaussianKernel

dogs = pd.read_csv('GFR-CKD.csv', index_col='ID Cane')
dogs.tail()
```

Li divido in pattern e etichette

```python
dogs_values = dogs.ix[:,1:]
dogs_labels = dogs.ix[:,0].values
```

Estraggo le prime due componenti principali

```python
from sklearn.preprocessing import StandardScaler
dogs_values_std = StandardScaler().fit_transform(dogs_values)

from sklearn.decomposition import PCA

pca_2d = PCA(n_components=2)

dogs_values_2d = pca_2d.fit_transform(dogs_values_std)
```

Visualizzo la proiezione del dataset nelle due componenti principali

```python
%matplotlib inline

import matplotlib.pyplot as plt

def gr_dataset(): 
    for lab, col in zip((0, 1),
                        ('blue', 'green')):
        plt.scatter(dogs_values_2d[dogs_labels==lab, 0],
                    dogs_values_2d[dogs_labels==lab, 1],
                    label=lab,
                    c=col)

gr_dataset()
```

Definisco due membership: una per l'insieme dei cani sani e una per quello dei cani malati

```python
def to_membership_values(labels, target):
    return [1 if l==target else 0 for l in labels]

mu = {}
for target in (0, 1):
    mu[target] = to_membership_values(dogs_labels, target)
```

La funzione che segue permette di visualizzare le curve di livello delle membership apprese, quella successiva definisce una nuvola di punti estratti a caso nella regione in cui mi interessa classificare.

```python
def gr_membership_contour(estimated_membership):
    x = np.arange(-4, 4, .1)
    y = np.arange(-4, 4, .1)
    X, Y = np.meshgrid(x, y)
    zs = np.array([estimated_membership((x, y))
                   for x,y in zip(np.ravel(X), np.ravel(Y))])
    Z = zs.reshape(X.shape)

    membership_contour = plt.contour(X, Y, Z,
                                     levels=(.1, .3, .5, .7, .8, .9, .95), colors='k')
    plt.clabel(membership_contour, inline=1)

def g(m):
    #return (-5 + np.random.random(2*m) * 10).reshape((m, 2))
    xs = -6 + np.random.random(m) * 12
    ys = -0.6 + np.random.random(m) * 1.4
    return zip(xs, ys)
```

Per una scelta arbitraria dei parametri dell'algoritmo di apprendimento, apprendiamo la membership all'insieme dei cani malati e visualizziamola

```python
c_0 = 10
sigma_0 = .5

estimated_membership_0, _ = possibility_learn(dogs_values_2d,
                                              mu[0],
                                              c=c_0,
                                              k=GaussianKernel(sigma_0),
                                              sample_generator=g)

gr_dataset()
gr_membership_contour(estimated_membership_0)

plt.ylim(-1, 1)
plt.show()
```

Facciamo lo stesso per i cani sani

```python
c_1 = 100
sigma_1 = .225

estimated_membership_1, _ = possibility_learn(dogs_values_2d,
                                              mu[1],
                                              c=c_1,
                                              k=GaussianKernel(sigma_1),
                                              sample_generator=g)

gr_dataset()
gr_membership_contour(estimated_membership_1)

plt.ylim(-1, 1)
plt.show()
```

Sovrapponiamo i due insiemi di curve di livello

```python
gr_dataset()
for e in (estimated_membership_0, estimated_membership_1):
    gr_membership_contour(e)

plt.ylim(-2, 2)
plt.show()
```

Attrezziamoci per rifare la visualizzazione con una mappa di calore

```python
def gr_membership_graded(estimated_membership, color_map, alpha=0.99):
    x = np.arange(-4, 4, .1)
    y = np.arange(-4, 4, .1)
    X, Y = np.meshgrid(x, y)
    zs = np.array([estimated_membership((x, y))
                   for x,y in zip(np.ravel(X), np.ravel(Y))])
    Z = zs.reshape(X.shape)
    plt.imshow(Z, interpolation='bilinear', origin='lower',
               cmap=color_map, alpha=0.59, extent=(-4, 4, -4, 4))
```

Visualizziamo in questa nuova modalità la membership all'insieme dei malati

```python
import matplotlib.cm as cm
gr_membership_graded(estimated_membership_0, cm.Blues, alpha=0.6)
gr_dataset()
plt.ylim(-2,2)
plt.show()
```

Stessa cosa per i sani

```python
gr_membership_graded(estimated_membership_1, cm.Greens, alpha=0.6)
gr_dataset()
plt.ylim(-2,2)
plt.show()
```

Sovrapponiamo le mappe di calore

```python
gr_membership_graded(estimated_membership_0, cm.Blues, alpha=0.6)
gr_membership_graded(estimated_membership_1, cm.Greens, alpha=0.6)
gr_dataset()
plt.ylim(-2,2)
plt.show()
```

Azzardiamo un classificatore che assegna ogni punto all'insieme per cui la membership è massima

```python
def classify(item):
    return sorted([(m(item), l)
                   for m, l in zip((estimated_membership_0,
                                    estimated_membership_1),
                              (0, 1))])[-1][1]
```

Come si comporterebbe questo classificatore sui dati che abbiamo usato?

```python
results = zip(range(len(dogs_values_2d)), map(classify, dogs_values_2d), dogs_labels)
```

```python
[r for r in results if r[1] != r[2]]
```

L'errore sarebbe nullo. Va però sottolineato che abbiamo usato tutti i punti a disposizione sia per addestrare il classificatore, sia per valutarne la bontà.

Se calcoliamo la differenza tra le membership otteniamo un indice che va da -1 (sano) a 1 (malato). La cella seguente visualizza la corrispondente mappa di colore evidenziando in bianco la zona "grigia".

```python
estimated_membership_both = lambda _: estimated_membership_0(_) - estimated_membership_1(_)
gr_membership_graded(estimated_membership_both, cm.coolwarm, alpha=0)
gr_dataset()
plt.ylim(-2,2)
plt.show()
```

Facciamo le cose un po' meglio. Dividiamo i dati a disposizione in training set (70% del totale) e test set (30% del totale). Addestriamo a partire dal training set con una 3-fold cross validation per la model selection. Ripetiamo tutto da capo per due volte.

(Nota, due volte è poco, e peraltro anche la model selection è farlocca perché i parametri sono fissati, ma così se vogliamo fare degli esperimenti più seri siamo già pronti)

Ripetiamo il tutto estraendo da 2 a 5 componenti principali e vediamo train e test error.

```python
def get_generator(d):
    return lambda m: (-4 + np.random.random(d*m) * 8).reshape((m, d))
    
def classify(item, est_mu, classes):
    return sorted([(l, est_mu[l](item)) for l in classes], key=lambda i:i[1], reverse=True)[0][0]

def holdout(data_values, data_labels, classes,
            train_percentage, cs, ks, generator_factory,
            classify, num_holdouts, log=True):
  
    n = len(data_values)
    
    performance_train = performance_test = 0.0
    
    trials = range(num_holdouts)
    if log:
        trials = log_progress(trials, every=1)
    
    for trial in trials:
    
        permuted_indices = np.random.permutation(n)
        train_indices = permuted_indices[:int(n*train_percentage)]
        test_indices = permuted_indices[int(n*train_percentage):]
    
        train_values = [data_values[i] for i in train_indices]
        test_values = [data_values[i] for i in test_indices]
    
        mu = {} 
        mu_train = {}
        mu_test = {}
    
        result = []
    
        for target in classes:
            mu[target] = to_membership_values(data_labels, target)
            mu_train[target] = [mu[target][i] for i in train_indices]
            mu_test[target] = [mu[target][i] for i in test_indices]
        
    
        labels_train = [data_labels[i] for i in train_indices]
        labels_test = [data_labels[i] for i in test_indices]
    
        est_mu = {}
        
        targets = classes
        if log:
            targets = log_progress(classes, every=1)
        
        for target in targets:
            est_mu[target] = model_selection_cv(train_values,
                                                mu_train[target],
                                                3,
                                                cs,
                                                ks,
                                                sample_generator=generator_factory(c),
                                                log=log)[2][0]

            if len(est_mu.values()) < len(classes):
                continue # at least one class not inferred
        
        results_train = zip(range(len(train_values)),
                            map(lambda item: classify(item, est_mu, classes),
                            train_values), labels_train)   
        performance_train += 1.0 * len([r for r in results_train if r[1] != r[2]]) / len(results_train)
    
        results_test = zip(range(len(test_values)),
                           map(lambda item: classify(item, est_mu, classes),
                           test_values), labels_test)  
    
        performance_test += 1.0 * len([r for r in results_test if r[1] != r[2]]) / len(results_test)
        
    return (performance_train/num_holdouts, performance_test/num_holdouts)
```

```python
cs = (10,)
sigmas = (.5,)

ks = map(GaussianKernel, sigmas)
classes = (0, 1)

train_percentage = 0.7
num_holdouts = 2

results_df = pd.DataFrame()

for c in log_progress(range(2, 6), every=1):
    pca = PCA(n_components=c)
    dogs_values_tr = pca.fit_transform(dogs_values_std)
    
    perf_train, perf_test = holdout(dogs_values_tr, dogs_labels,
                                    classes, train_percentage, cs, ks,
                                    get_generator, classify, num_holdouts,
                                    log=False)
    
    results_df = results_df.append({'components': c,
                                    'train': perf_train,
                                    'test': perf_test},
                      ignore_index=True)

results_df
```

Proviamo a lavorare con un solo attributo, il primo.

```python
x_dogs = dogs.ix[:,1].values
mu_dogs = dogs.ix[:,0].values

(min(x_dogs), max(x_dogs))
```

Senza pretesa, fissando a mano un po' dei gradi di libertà, la seguente cella apprende e visualizza la funzione di membership.

```python
from possibilearn.fuzzifiers import ExponentialFuzzifier

from possibilearn.kernel import LinearKernel
kernel = LinearKernel()
c = 5
g = lambda m: 4 * np.random.random(m)

estimated_membership, _ = possibility_learn([[i] for i in x_dogs],
                                            mu_dogs,
                                            c=c,
                                            k=kernel,
                                            sample_generator=g,
                                            fuzzifier=ExponentialFuzzifier(200))

x_test = np.arange(0, 4, .1)

g_estimated = plt.plot(x_test, map(estimated_membership, x_test))

g_dots = plt.scatter(x_dogs, [0] * len(x_dogs), c=1-np.array(mu_dogs), cmap='gray')

plt.xlim((0, 4))
plt.ylim((-0.1, 1.1))
plt.show()
```

```python
from ipywidgets import interact
import ipywidgets as widgets

from scipy.optimize import fsolve

kernel = LinearKernel()

def f(c, fuz):
    estimated_membership, _ = possibility_learn([[i] for i in x_dogs],
                                            mu_dogs,
                                            c=c,
                                            k=kernel,
                                            sample_generator=g,
                                            fuzzifier=fuz)

    x_test = np.arange(0, 4, .1)

    g_estimated = plt.plot(x_test, map(estimated_membership, x_test))

    g_dots = plt.scatter(x_dogs, [0] * len(x_dogs), c=1-np.array(mu_dogs), cmap='gray')

    low_y = 0.4
    hig_y = 0.6
    low_x = fsolve(lambda x: estimated_membership([x])-low_y, 2)[0]
    hig_x = fsolve(lambda x: estimated_membership([x])-hig_y, 2)[0]
    
    plt.plot([0, low_x, low_x], [low_y, low_y, -0.1])
    plt.plot([0, hig_x, hig_x], [hig_y, hig_y, -0.1])
    
    plt.xlim((0, 4))
    plt.ylim((-0.1, 1.1))
    plt.show()
    
interact(f, c=widgets.FloatSlider(min=0.1,max=30,step=.1,value=3.9),
            fuz=widgets.Dropdown(options=[LinearFuzzifier(),
                                          QuantileConstantPiecewiseFuzzifier(),
                                          QuantileLinearPiecewiseFuzzifier(),
                                          ExponentialFuzzifier(5),
                                          ExponentialFuzzifier(50),
                                          ExponentialFuzzifier(500)],
                                 value=LinearFuzzifier()));
```
