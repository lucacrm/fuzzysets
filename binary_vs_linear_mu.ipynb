```python
import numpy as np
import matplotlib.cm as cm
import pandas as pd
from muzzifiers import *
from possibilearn import *
from possibilearn.kernel import GaussianKernel
import support_vector_fuzzifier as svf

iris_df = pd.read_csv('iris.csv')

iris_df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
iris_df.dropna(how="all", inplace=True) # drops the empty line at file-end

iris_values = iris_df.iloc[:,:4].values #array dei soli valori
iris_labels = iris_df.iloc[:,4].values #array delle etichette
```

```python
out_file = open("binary_vs_linear_results.txt","w")
seed = np.random.randint(0,2**32)
sigmas = np.array([0.288, 0.250, 0.222])
cs = np.array([10**i for i in range(-2,3)])
```

```python
results = svf.iterate_tests(iris_values, iris_labels, cs, sigmas, 10, 
                            seed=seed, muzzifier=BinaryMuzzifier())
```

```python

out_file.write('Binary mus')
out_file.write('\n\n')
for k in results:
    out_file.write(k)
    out_file.write(': ')
    out_file.write(str(results[k]))
    out_file.write('\n')
    
out_file.write('\n\n\n')
```

```python
new_results = svf.iterate_tests(iris_values, iris_labels, cs, sigmas, 10, 
                                seed=seed, muzzifier=LinearMuzzifier())
```

```python
out_file.write('Linear mus')
out_file.write('\n\n')
for k in new_results:
    out_file.write(k)
    out_file.write(': ')
    out_file.write(str(new_results[k]))
    out_file.write('\n')
    
out_file.write('\n\n\n')
```

```python
out_file.close()
```
