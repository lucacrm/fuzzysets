```python
import numpy as np
import matplotlib.cm as cm
import pandas as pd
from muzzifiers import *
from possibilearn.fuzzifiers import *
from possibilearn.kernel import GaussianKernel
import support_vector_fuzzifier as svf

iris_df = pd.read_csv('iris.csv')

iris_df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
iris_df.dropna(how="all", inplace=True) # drops the empty line at file-end

iris_values = iris_df.iloc[:,:4].values #array dei soli valori
iris_labels = iris_df.iloc[:,4].values #array delle etichette
```

```python
def print_on_file(fl, d):
    for k in d:
        fl.write(k)
        fl.write(': ')
        fl.write(str(d[k]))
        fl.write('\n')
    
```

```python
out_file = open("fuzzifier-results.txt","w")
seed = np.random.randint(0,2**32)
sigmas = np.array([0.288, 0.250, 0.222])
cs = np.array([10**i for i in range(-2,3)])
for f in [CrispFuzzifier(), LinearFuzzifier(), QuantileConstantPiecewiseFuzzifier()]:
    out_file.write(f.name)
    out_file.write('\n\n')
    
    results = svf.iterate_tests(iris_values, iris_labels, cs, sigmas, 10, seed=seed,
                                muzzifier=BinaryMuzzifier(), fuzzifier=f)
    for k in results:
        out_file.write(k)
        out_file.write(': ')
        out_file.write(str(results[k]))
        out_file.write('\n')
    
    out_file.write('\n\n\n')
    
out_file.close()
```
