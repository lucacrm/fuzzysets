```python
import numpy as np
import matplotlib.cm as cm
import pandas as pd
from muzzifiers import *
from possibilearn.fuzzifiers import *
from possibilearn.kernel import GaussianKernel
import support_vector_fuzzifier as svf

source = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris_df = pd.read_csv(
    filepath_or_buffer=source,
    header=None,
    sep=',')

iris_df.columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid', 'class']
iris_df.dropna(how="all", inplace=True) # drops the empty line at file-end

iris_values = iris_df.iloc[:,:4].values #array dei soli valori
iris_labels = iris_df.iloc[:,4].values #array delle etichette
```

```python
out_file = open("different_fuzzifiers_results_2.txt","w")
sigmas = np.linspace(0.2,0.3,5)
cs = np.array([10**i for i in range(-1,3)])
```

```python
for (m,f) in [(m,f) for m in [BinaryMuzzifier(), LinearMuzzifier()] for f in [CrispFuzzifier(), LinearFuzzifier(),
                                               QuantileConstantPiecewiseFuzzifier()]]:
            
    n=f.name+'_'+m.name+'_2dim'
    results = svf.iterate_tests(iris_values, iris_labels, cs, sigmas, 10, dim=2, fuzzifier=f, muzzifier=m, name=n,
                               save_graph=True)
    out_file.write(svf.pretty_results(results))
    out_file.write('\n\n')
    
out_file.close()
```
